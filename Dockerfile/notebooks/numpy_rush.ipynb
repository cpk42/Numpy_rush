```python
import numpy as np
import matplotlib.pyplot as mpl
import matplotlib.image as img
```

```python
#question 1
def gen_arr(r, c):
    count = 0
    mtrx = np.ones(shape=(r, c))
    for i in range(r):
        for x in range(c):
            mtrx[i][x] = count
            count += 1

    return (mtrx)
    
matrx = gen_arr(10,10)
print(matrx)
```

```python
#question 2
seq = gen_arr(4, 3)
print(seq)
```

```python
#question 3
def gen_nsize(n):
    new = np.linspace(0, 1, n)
    return (new)

arr = gen_nsize(10)
print(arr)
```

```python
#question 4
def extract_arr(arr):
    if (arr.shape != (10, 12)):
        return (arr)
    else:
        grid = gen_arr(4, 4)
        x = 0
        for i in range(4):
            y = 8
            for j in range(4):
                grid[i][j] = arr[x][y]
                y += 1
            
            x += 1
        return (grid)

matrx1 = gen_arr(10, 12)
matrx2 = extract_arr(matrx1)
print(matrx2)
```

```python
#question 5
def bind_vectors(m, n):
    matrix = gen_nsize(n)
    tmp = matrix
    i = 1
    while i < m:
        tmp = np.vstack((matrix, tmp))
        i += 1
    
    return (tmp)
    

matrx3 = bind_vectors(320, 460)
mpl.imshow(matrx3)
```

```python
#Helper functions

#plot img
def plti(im, h=8, **kwargs):
    y = im.shape[0]
    x = im.shape[1]
    w = (y/x) * h
    mpl.imshow(im, interpolation="none", **kwargs)
    mpl.axis('off')

#adjust rgb channel value
def channel_adjust(channel, values):
    orig_size = channel.shape
    flat_channel = channel.flatten()
    adjusted = np.interp(
        flat_channel,
        np.linspace(0, 1, len(values)),
        values)

    return adjusted.reshape(orig_size)

#splits img into rgb channels
def channel_split(image):
    r = image[:, :, 0]
    g = image[:, :, 1]
    b = image[:, :, 2]
    return r, g, b

#merges rgb channel back to img matrix
def channel_merge(r, g, b):
    return np.stack([r, g, b], axis=2)

#masks a circular matrix to be used as a vignette in img processing
def circle_mask(m,n):
    mask = np.ones((m,n))
    for y in range(0,m):
        for x in range(0,n):
            mask[y][x] *= np.sin(np.pi*(x/n)) * np.sin(np.pi*(y/m))
    
    return (mask)

#creates a circle object to the size of the img
def circle(im):
    cir = circle_mask(im.shape[0], im.shape[1])
    for i in range(0, 4):
        im[:,:,i] *= cir
    return (im)
```

```python
#Part 2
def nashville(im):
    r, g, b = channel_split(im)
    r_b = channel_adjust(r, [0,.3, .75, 1])
    g_b = channel_adjust(g, [0,.7, 1])
    b_b = channel_adjust(b, [0, .7, .96, 1])
    res = channel_merge(r_b, g_b, b_b)
    plti(res)

def sutro(im):
    im = circle(im)
    r, g, b = channel_split(im)
    r_b = channel_adjust(r, [0, .75, 1])
    g_b = channel_adjust(g, [0, .6, 1])
    b_b = channel_adjust(b, [0, .6, 1])
    res = channel_merge(r_b, g_b, b_b)
    plti(res)

def earlybird(im):
    im = circle(im)
    r, g, b = channel_split(im)
    r_b = channel_adjust(r, [0, .85, 1])
    g_b = channel_adjust(g, [0, .55, .85, 1])
    b_b = channel_adjust(b, [0, .5, 1])
    res = channel_merge(r_b, g_b, b_b)
    plti(res)

def mayfair(im):
    r, g, b = channel_split(im)
    r_b = channel_adjust(r, [0, .4, .9, 1])
    g_b = channel_adjust(g, [0, .35, .95, 1])
    b_b = channel_adjust(b, [0, .45, .8, 1])
    res = channel_merge(r_b, g_b, b_b)
    plti(res)

#BONUS FILTERS

def gotham(im):
    plti(im)
    r, g, b = channel_split(im)
    r_b = channel_adjust(r, [0, 0.05, 0.1, 0.2, 0.3, 0.5, 0.7, 0.8, 0.9, 0.95, 1.0])
    b_b = np.clip(channel_adjust(b, [0, 0.047, 0.118, 0.251, 0.318, 0.392, 0.42, 0.439, 0.475, 0.561, 0.58, 0.627, 0.671, 0.733, 0.847, 0.925, 1]) + 0.03, 0, 1.0)
    im = channel_merge(r_b, g, b_b)
    plti(im)
    
def _1977(im):
    r, g, b = channel_split(im)
    r_b = channel_adjust(r, [0, .92, 1])
    g_b = channel_adjust(g, [0, .4, .95, 1])
    b_b = channel_adjust(b, [0, .6, .83, 1])
    res = channel_merge(r_b, g_b, b_b)
    plti(res)
```
